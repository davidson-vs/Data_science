# Avaliando base de booleanos

```python
import pandas as pd

tabela = pd.read_csv("../data/aluguel.csv", sep = ";")

print(tabela['Tipo'].value_counts())

# trazendo apenas os tipos que são apartamentos

selecao = tabela['Tipo'] == "Apartamento"
# print(selecao)
aps = tabela[selecao]

aps
```

```python
# Selecionando os tipos de casas
selecao = (tabela["Tipo"] == "Casa") | (tabela["Tipo"] == "Casa de Condomínio") | (tabela["Tipo"] == "Casa de Vila")
casas = tabela[selecao]
casas
```

```python
#Selecionando imóveis que tenham área entre 60 e 100 metros quadrados

selecao = (tabela["Area"] >= 60) & (tabela["Area"] <= 100)
intervalo_areas = tabela[selecao]
intervalo_areas
#para mostrar o numero de dados existente => intervalo_areas.shape[0]
```
