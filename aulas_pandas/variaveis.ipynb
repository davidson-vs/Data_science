# Criando variaveis

```python
import pandas as pd

dados = pd.read_csv("../data/aluguel_residencial.csv", sep= ";")

dados["Valor Bruto"] = dados["Valor"] + dados["Condominio"] + dados["IPTU"]

dados["Valor m2"]  = round(dados["Valor"]/dados["Area"],2)

dados["Valor Bruto m2"]  = round(dados["Valor Bruto"]/dados["Area"],2)

casas = ['Casa', "Casa de Vila", "Casa de Condomínio"]
dados["Agregado"]  = dados["Tipo"].apply( lambda x : "Casa" if x in casas else "Apartamento")


dados.head(15)
```

## Como remover variaveis

```python
dt_aux = pd.DataFrame(data= dados[["Agregado","Valor Bruto", "Valor m2", "Valor Bruto m2"]])

#uma das formas
del dt_aux["Agregado"]

#outra forma

dt_aux.pop("Valor Bruto m2")

# Forma mais usada 

dados.drop(["Valor Bruto", "Valor Bruto m2"], axis= 1, inplace= True)

dados.head(10)

dados.to_csv("../data/aluguel_residencial.csv", sep= ";")
```
